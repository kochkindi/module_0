```python
Данный алгоритм можно улучшить, используя БИНАРНЫЙ поиск. Суть его состоит в том, что мы загадываем число в середине диапазона
и исключаем половину оставшихся чисел.
```

```python
import numpy as np
def search_function():
    '''Устанавливаем любое random число, которое нам нужно угадать, генерируем список с нашим выбранным диапазоном [1,100] 
и определяем индекс минимального и максимального элемента в этом списке.
    Функция принимает эти параметры и по итогу возвращает загаданное число и число попыток'''
    count = 0 # счетчик попыток
    number = np.random.randint(1,101) # загадали число (число можно загадать любым способом)
    our_list = [i for i in range(1,101)] # генерируем лист с выбранным диапазоном
    low_index = 0 
    high_index = len(our_list)-1
    
    while low_index<=high_index: # цикл, где мы будем поочередно проводить бинарный поиск
        count += 1 
        mid_index = int((low_index + high_index)/2) # находим индекс числа в середине диапазона
        guess = our_list[mid_index] # находим число в середине диапазона по индексу
        if guess == number: # проверяем, не угадали ли мы число, если да - выходим из цикла
            return print (f"Загаданное число: {number}. Вы угадали число {guess} за {count} попытки/ток.")
        if guess > number: 
            print(f"{count} попытка: предполагаемое число {guess} больше загаданного.")
            high_index = mid_index-1  # в таком случае устанавливаем новый индекс максимального элемента, сужаем поиск
        else:
            print(f"{count} попытка: предполагаемое число {guess} меньше загаданного.")
            low_index=mid_index+1 # в таком случае устанавливаем новый индекс минимального элемента, сужаем поиск
    return None # если индекс максимального элемента меньше минимального
    
    
search_function () 
```

```python
Мы выяснили, что алгоритм, представленный выше более эффективен. 
В общем случае для списка из n элементов бинарный поиск выполняется за log2(n) шагов, тогда как простой поиск займет n шагов.
```
